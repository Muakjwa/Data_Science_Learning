# LazyPredict

## 이제 model을 제작해야하는데 Lazypredict를 사용할 것이다.
따라서 먼저 train_test_split으로 train,test 를 분리해준다.

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25, random_state= 30)
```

```python
print(X_train.shape, X_test.shape)
```

```python
clf = LazyClassifier(predictions=True)
model_classification, predictions_classification = clf.fit(X_train,X_test,y_train,y_test)
print(model_classification)
```

정확도가 0.91이상이 나오지 않아 StandardScaler을 사용해보았지만 결과는 같았다.
Kaggle Leaderboard에서는 99퍼 이상이 다수를 차지하고 있었다.

```python
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
model_classification, predictions_classification = clf.fit(X_train_st,X_test_st,y_train,y_test)
print(model_classification.head)
```

```python
parameters = {'booster':['gbtree'], 'n_estimators':[130], 'learning_rate':[0.2], 'reg_lambda':[2]
              ,'importance_type':['gain'], 'reg_alpha':[0], 'objective':['binary:logistic'], 'gamma':[0]}

xgbc = XGBClassifier()
grid = GridSearchCV(xgbc, param_grid = parameters, cv = 3)

grid.fit(X,y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
xgbc = XGBClassifier(booster='gbtree'
                   ,n_estimators=130
                   ,learning_rate=0.2
                   ,reg_lambda=2
                   ,random_state=30)

xgbc.fit(X_train,y_train)

print(xgbc.score(X_test,y_test))
```

```python
test.columns = X.columns
```
